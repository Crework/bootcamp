### List Comprehensions

If I tell you to create a list of **whole numbers upto 10** you'll probably write the following code:-

```python
whole_number_till_10 = []

for i in range(10):
    whole_number_till_10.append(i)
```

Or maybe if I say to create a list of **even numbers upto 10** you'll probably write the following code:-

```python
even_number_till_10 = []

for i in range(10):
    if i % 2 == 0:
        even_number_till_10.append(i)
```

While the above codes are correct, they aren't exactly **tidy**. List comprehension is a great utility by which you can create simple list in just one single line. So the above two becomes:-

```python
whole_number_till_10 = [i for i in range(10)]
even_number_till_10 = [i for i in range(10) if i % 2 == 0]
```

### Lambda Functions




Lambda functions are utility by which you can create functions in a single line. So if you have a simple function you can wrap that as a lambda function. Few usecases of lambda functions can be when you try to apply a function over a whole column in the dataframe, itertools, etc.

Let's take following function as the example:-
```python
def func(a):
    return a + 5

print(func(6))
```

We can create the lambda function of the above as:-

```python
func = lambda a : a + 5

print(func(6))
```

### Ternary Operator

Ternary Operator is basically writing if-else statement in a single line. Shorter the code the more we love it. One thing to note is that you can only wrap simple if-else blocks as ternary operator. In Python, the format for ternary operator is:-
```
result-if-true if <condition> else result-if-false
```

In terms of usage you can use it at various places like print statement, list comprehensions, lambda functions etc. 

Let's take following as the example:-
```python
a = 5
if a % 2 == 0:
    a += 1
else:
    a += 2
```

We can create the above as a one-liner as:-
```python
a = 5
a += 1 if a % 2 == 0 else 2
```

### map(), filter(), reduce()

Sometimes you'll feel like there are operation that you wanna apply over an iterator or a list. Some of the common itertools you can use are:-

* **map():** Used to apply a function over the elements of an iterator.
* **filter():** Used to filter out the element that satisfy a condition tested via a function, the element is kept if the function returns true.
* **reduce():** Used to accumalate the results of the iterator elements when an operation is applied.

A really cool infographic to understand the application of the above three is:-
map-filter-reduce-in-emoji-1.png


Let's take an example to understand the working of the above 3. In order to understand there working a bit better:-
```python
l = [i for i in range(10)]
print(list(map(lambda x: x**2, l)))
```

The above code takes a list as input and applies square over each element. Notice we are converting it to list because map() returns a map object which is an iterator.
```python
l = [i for i in range(10)]
print(list(filter(lambda x: True if x % 2 == 0 else False, l)))
```

The above code takes a list as input and selects out only numbers which are even.
```python
from functools import reduce

l = [i for i in range(10)]
print(reduce(lambda a,b: a + b, l))
```

The above code takes a list as input and returns the sum of list. In the lambda function we explained how the numbers have to be accumalated i.e. sum in this case.

For more info you can refer: https://book.pythontips.com/en/latest/map_filter.html

## Classes and Object

Classes are the lifeline of Python, everything you see in Python is a class. From intergers to function everything is a class. Objects are defined as different instances to the class that can access the attributes and methods of the class. Format to create a class is as follows:-

```python
class Crework:
    def __init__(self, attr1, attr2):
        self.attr1 = attr1
        self.attr2 = attr2
    
    def method1(self):
        pass
    
    def method2(self):
        pass

batch_1 = Crework("BWU", 1)
batch_2 = Crework("BWU", 2)
batch_3 = Crework("BWU", 3)
```
Cool isn't it? In the above code we created a class name **Crework** with 2 class attributes i.e. attr1 and attr2 alson with 2 methods i.e. method1 and method2. After that we created 3 objects of Crework and note that all of them have different value of attr2. 

__init__ is a constructor class called each time an object is created, the function parameters to __init__ are also the arguments to it while creating objects. If we have to explain about **self** it's basically a parameter that can be used to access all class methods and attributes, similar to this pointer in C++ and Java.

## Dot Product

## Reshape and Resize 

Most of us are familiar with reshape utilities in numpy ie. **reshape** and **resize**. Both of them can be used to resize a particular array to another shape. But how do these 2 differ?

While they both do the same thing the difference lies in the way they do so. To write it down for you:-

* **reshape()** will return the reshaped array without changing the shape of the original.
* **resize()** will reshape the given array given array inplace meaning it's modifying the original array itself.

```python
a = np.array([i for i in range(10)])

# Reshape returns a array
b = a.reshape(2,5)

# Resize modifies original array
a.resize(2,5)
```

**NOTE:** There is a difference between reshape and transpose, reshape will flatten the array first and then recreate the array of the given shape by default. These ordering can be changed by **order** parameter.

# Combined Question

Below we have the data for 'films', their release 'years', and their 'characters'. Notice how the case of the 'characters' is inconsistent in the dataframe and the dictionary.

````python
import pandas as pd
````

````python
data = pd.DataFrame({'Characters':['Mr pink','Zoe Bell', 'Vincenzo Coccotti', 'Calvin Candie', 'Ordell Robbie', 'richie gecko', 'Shosanna', 'Oswaldo Mobray', 'Captain Koons', 'Mallory Knox'], 
                     'Years':[1992, 2007, 1993, 2012, 1997, 1996, 2009, 2015, 1994, 1994]})
data
````

````python
films = {'mr pink': 'Reservoir Dogs', 'zoe bell': 'Death Proof', 'Vincenzo Coccotti':'True Romance', 'Calvin Candie':'Django Unchained', 'Ordell Robbie':'Jackie Brown', 'Richie Gecko':'From Dusk Till Dawn', 'Shosanna':'Inglourious Basterds', 'Oswaldo Mobray':'The Hateful Eight', 'Captain Koons':'Pulp Fiction', 'Mallory Knox':'Natural Born Killers'}
films
````

We can use the map method to combine the character data with their films. But the problem here is that some of the characters are capitalized and others are not.

````python
data['Films'] = data['Characters'].map(films)
data
````


So when we did the mapping, as you can see we have some missing values.

Thus, we need to convert each value from the character list to lowercase using the `str.lower()` Series method.

````python
lowercased = data['Characters'].str.lower()
data['Films'] = lowercased.map(films)
data
````

This however only now matched with the first two films where the character key was also in lowercase. There is no simple function to convert the keys in the dict to lowercase, so instead we need to write a function to iterate over the dict to create a new dict with all keys turned to lower case.

````python
def lower_dict(d):
   new_dict = dict((k.lower(), v) for k, v in d.items())
   return new_dict

lower_films = lower_dict(films)
lower_chars = data['Characters'].str.lower()
data['Films'] = lower_chars.map(lower_films)
data
````


We have now successfully mapped the character data with the films.
